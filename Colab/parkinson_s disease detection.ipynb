```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn import svm
```

```python
parkinson_data=pd.read_csv('/content/Parkinsson disease.csv')
```

0---->parkinson's neagtive---healthy
1--->parkinson's positive---unhealthy

```python
#grouping the data based on status
parkinson_data.groupby('status').mean()
```

seperating features and target

```python
x=parkinson_data.drop(['name','status'],axis=1)
y=parkinson_data['status']
```

```python
print(x)
```

```python
print(y)
```

splitting into training and testing data

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

data standarization

```python
scaler=StandardScaler()
```

```python
x_train=scaler.fit_transform(x_train)
```

```python
x_test=scaler.transform(x_test)
```

```python
print(x_train)
```

```python
print(x_test)
```

model training

support vector machine

```python
model=svm.SVC(kernel='linear')
```

```python
model.fit(x_train,y_train)
```

model evaluation

```python
x_train_predict=model.predict(x_train)
x_train_accuracy=accuracy_score(x_train_predict,y_train)
print(x_train_accuracy)
```

```python
#on testing
x_test_predict=model.predict(x_test)
x_test_accuracy=accuracy_score(x_test_predict,y_test)
print(x_test_accuracy)
```
