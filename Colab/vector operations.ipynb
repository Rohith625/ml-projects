addition of two vectors

```python
vector_1=np.asarray([0,0,2,3])

vector_2=np.asarray([0,0,3,-2])
sum=vector_1 + vector_2
print(sum)
```

```python
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='r')
plt.quiver(0,0,3,-2,scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,5,1,scale_units='xy',angles='xy',scale=1,color='y')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

subtraction of two vectors

```python
vector_1=np.asarray([0,0,2,3])

vector_2=np.asarray([0,0,3,-2])
diff=vector_1 - vector_2
print(diff)
```

```python
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='r')
plt.quiver(0,0,3,-2,scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,-1,5,scale_units='xy',angles='xy',scale=1,color='y')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

dot product of two vectors

```python
a=np.array([2,3])
b=np.array([4,5])

a_dot_b=np.dot(a,b)
print(a_dot_b)
```

```python
a=np.array([5,6,7])
b=np.array([3,4,9])
a_dot_b1=np.dot(a,b)
print(a_dot_b1)
```

cross product of two vectors

```python
a=np.array([2,3])
b=np.array([4,5])

a_cross_b=np.cross(a,b)
print(a_cross_b)
```

```python
c=np.array([2,3,4])
d=np.array([5,10,30])
c_cross_d=np.cross(c,d)
print(c_cross_d)
```

projection of 'a' vector of 'b' vector

vector-projection.jpg

```python
a=np.array([2,5])
b=np.array([8,-6])
#magnitude of b vector
magnitude_b=np.sqrt(np.sum(b**2))
#projection of a on b
proj_of_a_on_b=(np.dot(a,b)/magnitude_b**2)*b
print(proj_of_a_on_b)
```

```python
c=np.array([2,3,4])
d=np.array([5,6,7])
#magnitude of vector
magni_d=np.sqrt(np.sum(d**2))
#projection of c on d
proj_of_c_on_d=(np.dot(c,d)/magni_d**2)*d
print(proj_of_c_on_d)
```
