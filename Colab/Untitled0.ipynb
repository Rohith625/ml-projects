Matplotlib


*   useful for making plots



```python
#import matplotlib library
import matplotlib.pyplot as plt
```

```python
#import numpy
import numpy as np
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

plotting the data

```python
#sine wave
plt.plot(x,y)
plt.show()
```

```python
#cos wave
plt.plot(x,z)
plt.show()
```

```python
#adding titile x axis and y axis label
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sinevalue')
plt.title('sine wave')
plt.show()
```

```python
#parabola
x=np.linspace(-10,10,20)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'r+')
plt.show()
```

```python
plt.plot(x,y,'g.')
plt.show()
```

```python
plt.plot(x,y,'rx')
plt.show()
```

```python
x=np.linspace(-5,+5,50)
plt.plot(x,np.sin(x),'g-')
plt.plot(x,np.cos(x),'r*')
plt.show()
```

bar plot

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
animals=['cat','dog','rat','lion','bear']
zoo=[100,50,40,30,70]
ax.bar(animals,zoo)
plt.xlabel('animals')
plt.ylabel('no.of animals')
plt.show()
```

pie chart

```python
fig1=plt.figure()
ax=fig1.add_axes([0,0,1,1])
animals=['cat','dog','rat','lion','bear']
zoo=[100,50,40,30,70]
ax.pie(zoo,labels=animals,autopct='%1.1f%%')
plt.show()
```

scatter plot

```python
from matplotlib import color_sequences
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
fig2=plt.figure()
ax=fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='r')
plt.show()
```

3D scatter plot

```python
fig3=plt.figure()
ax=plt.axes(projection='3d')
z=20 * np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,cmap='Blues')
plt.show()
```
