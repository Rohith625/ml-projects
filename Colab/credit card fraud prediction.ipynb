```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_data=pd.read_csv('/content/credit_data.csv')
```

```python
credit_data.isnull().sum()
```

```python
credit_data['Class'].value_counts()
```

```python
credit_data=credit_data.dropna()
```

```python
credit_data.shape
```

```python
credit_data.isnull().sum()
```

0 ----> Legit transaction
1 ----> fraud transaction

```python
credit_data['Class'].value_counts()
```

this is highly imbalanced dataset

```python
#sepearting the data for analysis
legit=credit_data[credit_data.Class==0]
fraud=credit_data[credit_data.Class==1]
```

```python
print(legit.shape,fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the values for both transactions
credit_data.groupby('Class').mean()
```

dealing with imbalanced data

using undersampling

building a sample dataset containing a similar distribution of normal and fraud transactions

```python
legit_sample=legit.sample(n=300)
```

```python
#concating two data frames
new_data=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_data.head()
```

```python
new_data.shape
```

```python
new_data['Class'].value_counts()
```

```python
new_data.groupby('Class').mean()
```

splitting the data and target

```python
x=new_data.drop(['Class'],axis=1)
y=new_data['Class']
```

```python
print(x)
```

```python
print(y)
```

split into training and testing

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

model training

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

evaluation

```python
#on training data
x_train_predict=model.predict(x_train)
x_train_accuracy=accuracy_score(x_train_predict,y_train)
print(x_train_accuracy)
```

```python
#on test data
x_test_predict=model.predict(x_test)
x_test_accuracy=accuracy_score(x_test_predict,y_test)
print(x_test_accuracy)
```
