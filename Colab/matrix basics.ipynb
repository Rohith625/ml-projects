matrix operations

matrix addition

```python
vector_1=np.array([[1,2],[4,5]])
vector_2=np.array([[6,7],[8,9]])
sum=vector_1 + vector_2
print(sum)
```

```python
vector_1=np.array([[1,2],[4,5]])
vector_2=np.array([[6,7],[8,9]])
c=np.add(vector_1,vector_2)
print(c)
```

```python
a=np.random.randint(100,size=(3,4))
print(a)
b=np.random.randint(150,size=(3,4))
print(b)
c=a+b
print(c)
```

```python
a=np.random.randint(100,size=(3,4))
print(a)
b=np.random.randint(150,size=(3,4))
print(b)
c=np.add(a,b)
print(c)
```

matrix subtraction

```python
vector_1=np.array([[1,2],[4,5]])
vector_2=np.array([[6,7],[8,9]])
diff=vector_1 - vector_2
print(diff)
```

```python
vector_1=np.array([[1,2],[4,5]])
vector_2=np.array([[6,7],[8,9]])
diff=np.subtract(vector_1,vector_2)
print(diff)
```

```python
a=np.random.randint(100,size=(3,4))
print(a)
b=np.random.randint(150,size=(3,4))
print(b)
c=b-a
print(c)
```

```python
a=np.random.randint(100,size=(3,4))
print(a)
b=np.random.randint(150,size=(3,4))
print(b)
c=np.subtract(b,a)
print(c)
```

multiplying a matrix with scalar

```python
vector_1=np.array([[1,2],[4,5]])
vector_2= 2*vector_1
print(vector_2)
```

```python
vector_1=np.array([[1,2],[4,5]])
x=2
product=np.multiply(x,vector_1)
print(product)
```

```python
a=np.random.randint(10,size=(3,4))
print(a)
b=2*a
print(b)
```

```python
a=np.random.randint(10,size=(3,4))
print(a)
b=2
c=np.multiply(a,b)
print(c)
```

multiplying matrices

```python
vector_1=np.array([[1,2],[4,5]])
vector_2=np.array([[6,7],[8,9]])
product=np.dot(vector_1,vector_2)
print(product)
print(product.shape)
```

```python
a=np.random.randint(5,size=(3,4))
print(a)
b=np.random.randint(5,size=(4,5))
print(b)
c=np.dot(a,b)
print(c)
print(c.shape)
```
