this is highly imbalanced datset

```python
#seperating the legit and fraud transcations
legit=dataset[dataset.Class==0]
fraud=dataset[dataset.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

undersampling

build a sample dataset containing similar distribution of legit and fraud

```python
legit_sample=legit.sample(n=492)
```

```python
print(legit_sample.shape)
```

concateneate two dataframes

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
print(new_dataset.shape)
```
