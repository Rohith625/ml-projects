**Gradient descent**

It is an optimization alogrithm used for minimizing the loss function in various machine learning alogrithms.It is used for updating the parameters of the learning model.


w=w-L*dw

b=b-L*db

**Learning rate**(L)

It is a tuning parameter in optimization alogirthm that determines the step size at each iteration while moving towards minimum of loss function

Linear Regression

```python
class Linear_Regression():

  #initiating the parameters (learning rate and no of iterations )
  def __init__(self,learning_rate,no_of_iterations):
    self.learning_rate=learning_rate
    self.no_of_iterations=no_of_iterations

  def fit(self,x,y):

    #no of training examples & no of features
    self.m,self.n=x.shape    #no of rows and columns

    #initating the weight and bias
    self.w=np.zeros(self.n)
    self.b=0
    self.x=x
    self.y=y

    #implementing gradient descent
    for i in range(self.no_of_iterations):
      self.update_weights()

  def update_weights(self):
    y_prediction=self.predict(self.x)

    #calculate gradients
    dw= -(2*(self.x.T).dot(self.y-y_prediction))/self.m

    db= -2 *np.sum(self.y-y_prediction)/self.m

    #update the weights
    self.w=self.w-self.learning_rate*dw
    self.b=self.b-self.learning_rate*db

  def predict(self,x):
    return x.dot(self.w) + self.b
```
