# Data Visualization

```python
plt.style.use('ggplot')
plt.figure(figsize=(7,7))
count1 = df[df['Attrition'] == 'Yes'].groupby('BusinessTravel')['Attrition'].count().reset_index()
plot1=plt.bar(count1['BusinessTravel'],count1['Attrition'],color='#886655',width=0.5)
plt.bar_label(plot1, label_type='edge', color='r', fontsize=15)
plt.show()
```

```python
count2 = df[df['Attrition']=='Yes'].groupby('Department')['Attrition'].count().reset_index()
plot2=plt.bar(count2['Department'],count2['Attrition'],color=['#888888','#884466','#885566'],width=0.5)
plt.bar_label(plot2,label_type='edge', color='r', fontsize=15)
plt.show()
```

```python
count3 = df[df['Attrition'] == 'Yes'].groupby('EducationField')['Attrition'].count().reset_index()
plt.pie(count3['Attrition'],labels=count3['EducationField'],autopct='%1.1f%%',shadow=True,startangle=45,labeldistance=1.1)
plt.show()
```

```python
count4 = df[df['Attrition'] == 'Yes'].groupby(['Gender', 'OverTime'])['Attrition'].count().reset_index()
plot4 = sns.barplot(x='Gender', y='Attrition', hue='OverTime', data=count4)
for p in plot4.patches:
    plot4.annotate(format(p.get_height(), '.0f'),
                   (p.get_x() + p.get_width() / 2., p.get_height()),
                   ha = 'center', va = 'center',
                   xytext = (0, 9),
                   textcoords = 'offset points')
plt.show()
```

```python
colors = ['#e6e5e5','#9b9797','#858080','#4e4b4b']
count5 = df[df['Attrition'] == 'Yes'].groupby('MaritalStatus')['Attrition'].count().reset_index()
plt.pie(count5['Attrition'],labels=count5['MaritalStatus'],colors=colors,autopct='%.1f%%',labeldistance=1.1,
        pctdistance=0.8,startangle=40,textprops={'color':'#800000'})
center_circle=plt.Circle((0,0),0.6,color='White')
fig=plt.gcf()
fig.gca().add_artist(center_circle)
plt.show()
```

```python
plt.figure(figsize=(22,10))
count6=df[df['Attrition'] == 'Yes'].groupby('JobRole')['Attrition'].count().reset_index()
plot6=plt.bar(count6['JobRole'],count6['Attrition'],width=0.5,color='b',align='center')
plt.xticks(color='r')
plt.yticks(color='r')
plt.bar_label(plot6,color='r',fontsize=12,label_type='edge')
plt.show()
```

```python
plt.figure(figsize=(8,8))
count7=df[df['Attrition'] == 'Yes'].groupby(['JobSatisfaction','JobLevel'])['Attrition'].count().reset_index()
plot7 = sns.barplot(x=count7['JobLevel'],y=count7['Attrition'],hue=count7['JobSatisfaction'])
for p in plot7.patches:
    plot7.annotate(format(p.get_height(), '.0f'),
                   (p.get_x() + p.get_width() / 2., p.get_height()),
                   ha = 'center', va = 'center',
                   xytext = (0, 9),
                   textcoords = 'offset points')
plt.show()
```

```python
#taking some random samples in df
df1 = df.sample(n=200,random_state=2)
plt.scatter(df1['Age'],df1['MonthlyIncome'])
x = df['Age']
y = df['MonthlyIncome']
m, c = np.polyfit(x, y, 1)

# Adding line equation
plt.plot(x, m*x + c, color='blue', label=f'y = {m:.2f}x + {c:.2f}',linestyle='--')
plt.xlabel('Age')
plt.ylabel('Monthly Income')
plt.legend()
plt.show()
```

```python
count8=df[df['Attrition'] =='Yes'].groupby('Department')['TotalWorkingYears'].mean().reset_index()
count9=df[df['Attrition'] =='Yes'].groupby('Department')['PercentSalaryHike'].mean().reset_index()
count10 = df[df['Attrition']=='Yes'].groupby('Department')['YearsAtCompany'].mean().reset_index()
plt.plot(count8['Department'],count8['TotalWorkingYears'])
plt.plot(count9['Department'],count9['PercentSalaryHike'])
plt.plot(count10['Department'],count10['YearsAtCompany'])
plt.show()
```
